```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt
#matplotlib.rcParams['figure.figsize'] = (6,6)
import scipy, scipy.stats
```

```python
# Prior probability of success of a single Bernoulli experiment
alpha = 7 # think of alpha-1 as a pseudo-count for x=1
beta = 9  # think of beta-1  as a pseudo-count for x=0

# generate grid
x = scipy.linspace(0,1,100)

plt.figure(figsize=(10,6))

# evaluate the beta pdf
betapdf = scipy.stats.beta.pdf(x, alpha, beta)
plt.plot(x,betapdf)
plt.xlabel('$\mu$')
plt.ylabel('$p(\mu | \alpha, \beta)$')
```

```python
N = 15 # number of coin flips
heads = 10 # number of "heads"
tails = N - heads

# posterior
alpha_new = alpha + heads
beta_new = beta + tails
```

```python
# plot the beta prior and posterior
plt.figure(figsize=(10,6))
betapdf_prior = scipy.stats.beta.pdf(x, alpha, beta)
plt.plot(x,betapdf_prior)
betapdf_posterior = scipy.stats.beta.pdf(x, alpha_new, beta_new)
plt.plot(x,betapdf_posterior)
plt.legend(['Prior', 'Posterior'])
```
