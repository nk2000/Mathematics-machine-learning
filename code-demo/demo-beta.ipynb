```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt
#matplotlib.rcParams['figure.figsize'] = (6,6)
import scipy, scipy.stats
import subprocess
import shlex
```

```python
# Probability of success of a single Bernoulli experiment
alpha = [0.5, 1, 2, 4, 5]
beta = [0.5, 1, 0.3, 10, 1]

# generate integer grid
x = scipy.linspace(0,1,100)

plt.figure(figsize=(10,6))
for i in range(len(alpha)):

    # evaluate the beta pdf
    betapdf = scipy.stats.beta.pdf(x, alpha[i], beta[i])
    plt.plot(x,betapdf)
    
    
plt.xlabel('$\mu$')
plt.ylabel('$p(\mu|a,b)$')
plt.legend(['$a=0.5=b$', '$a=1=b$', '$a=2=b$', '$a=4, b = 10$', '$a=5, b = 1$'])
plt.axis('tight')
plt.savefig('aaa.pdf')
```

```python
# print figure
fname='beta_example.pdf'
plt.savefig(fname,bbox_inches='tight')
proc=subprocess.Popen(shlex.split('lpr {f}'.format(f=fname)))
```
