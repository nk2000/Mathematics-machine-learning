```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt
#matplotlib.rcParams['figure.figsize'] = (6,6)
import scipy, scipy.stats
import subprocess
import shlex
```

```python
# Number of "coin-flip" experiments
N = 15

# Probability of success of a single Bernoulli experiment
mu = [0.1, 0.4, 0.75]
col = ['r', 'b', 'g']

# generate integer grid
x = scipy.linspace(0,N,N+1)

plt.figure(figsize=(10,6))
for i in range(len(mu)):

    # evaluate the binomial pdf
    binpdf = scipy.stats.binom.pmf(x,N,mu[i])

    # plot
    plt.bar(x-0.5, binpdf, color=col[i], alpha = 0.25);
    plt.legend(['$\mu=0.1$', '$\mu=0.4$', '$\mu=0.75$'])
    plt.xlabel('Number $m$ of observations $x=1$ in $N=15$ experiments')
    plt.ylabel('$p(m)$')
    plt.axis('tight')
```

```python
# print figure
fname='binomial_example.pdf'
plt.savefig(fname,bbox_inches='tight')
proc=subprocess.Popen(shlex.split('lpr {f}'.format(f=fname)))
```
